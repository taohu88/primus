```python
import pandas as pd
import category_encoders as ce
import numpy as np
```

```python
df = pd.DataFrame({
        'name': ['The Dude', 'Walter', 'Donny', 'The Stranger', 'Brandt', 'Bunny'],
        'haircolor': ['brown', np.nan, 'brown', 'silver', 'blonde', 'blonde'],
        'gender': ['male', 'male', 'male', 'male', 'male', 'female'],
        'drink': ['caucasian', 'beer', 'beer', 'sasparilla', 'unknown', 'unknown'],
        'age': [48, 49, 45, 63, 40, 23] 
    },
    columns=['name', 'haircolor', 'gender', 'drink', 'age']
)

test_df = pd.DataFrame({
        'name': ['The Dude'],
        'haircolor': ['black'],
        'gender': ['male'],
        'drink': ['caucasian'],
        'age': [48] 
    },
    columns=['name', 'haircolor', 'gender', 'drink', 'age']
)
```

```python
encoder = ce.OneHotEncoder(cols=['haircolor']) #, handle_unknown='return_nan')
df_onehot = encoder.fit_transform(df)
df_onehot
```
